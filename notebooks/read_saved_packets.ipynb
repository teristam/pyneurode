```python
import pickle
import os
```

```python
packets = []
with open('../data/M7_2022-07-16_17-17-33_test1/packets_new.pkl', 'rb') as f:
    while True:
        try:
            msg = pickle.load(f)
            if type(msg) is list:
                packets.extend(msg)
            else:
                packets.append(msg)
        except EOFError:
            break

print(f'Total packets: {len(packets)}')
print(f'dtypes: {set(m.dtype for m in packets)}')
```

```python
# shorten packets for testing
with open('../data/M7_2022-07-16_17-17-33_test1/packets_new.pkl', 'rb') as f, open('../examples/spikes/test_packets_short.pkl', 'wb') as f_out:
    for i in range(10000):
        try:
            msg = pickle.load(f)
            pickle.dump(msg, f_out)
        except EOFError:
            break
```
