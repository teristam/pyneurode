```python
import pickle
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
```

```python

def loadData(filename):
    data = []
    with open(filename,'rb') as f:
        while True:
            try:
                d = pickle.load(f)
                data.append(d.data)
            except EOFError:
                break
    df = pd.concat(data)
    return df
```

```python
df = loadData(f'../data/df_sort_ch256_cy.pkl')
df['process_time'] = (df.sort_timestamps - df.acq_timestamps)*1000
plt.figure()
plt.plot(df.process_time.values)
print(df.process_time.iloc[50000:60000].mean())
```

```python
df = loadData(f'../data/df_sort_ch256_cy.pkl')
plt.figure(figsize=(8,5))
df['process_time'] = (df.sort_timestamps - df.acq_timestamps)*1000
df_sel = df.iloc[50000:60000]
sns.set_context('talk',font_scale=1.2)
ax = sns.histplot(x='process_time',binwidth=10, data=df_sel)
ax.set(xlabel='Latency (ms)')
plt.savefig('figures/latency_ch256_cy.png',dpi=300,bbox_inches='tight')
```
